```python
#------------FIRST STEP---------------------
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares"

r= requests.get(url)

if r.status_code == 200:
    text = r.text
    BeautifulSoup = BeautifulSoup(text)
    tableclass = "table table-striped table-responsive-md"
    tablebody = BeautifulSoup.find(class_ = tableclass)
    tablerows = tablebody.findAll("tr")
    tablerows
else:
    print("request failed!")
    
```

```python
#-------------- THIRD STEP ------------------------

import requests
import pandas as pd
from bs4 import BeautifulSoup
url = 'https://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html'
r = requests.get(url)


text = r.content
beautifulsoup = BeautifulSoup(text,'html5lib')
tableclass = 'pre'
new_table1 = beautifulsoup.find(tableclass)
#new_table1 = tablebody.findAll("tr")
#new_table1

    

```

```python

#get the text in the file
string = new_table1.get_text()

#Split each line and put in an array
lines = string.split('\n')

#Remove the first 11 unnecessary lines to make the file read the list
lines.pop(0);
lines.pop(0);
lines.pop(0);
lines.pop(0);
lines.pop(0);
lines.pop(0);
lines.pop(0);
lines.pop(0);
lines.pop(0);
lines.pop(0);
lines.pop(0);

#Remove the last three useless lines
lines.pop(len(lines)-1)
lines.pop(len(lines)-1)
lines.pop(len(lines)-1)

#Make the table
new_table1 = pd.DataFrame(columns=['start_date', 'start_time', 'end_date', 'end_time', 'start_frequency', 'end_frequency', 'flare_location', 'flare_region','flare_classification', 'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed'], index = range(0,len(lines)))

#Iterate through each line to put values in the table
row = 0
for line in lines:
    col = 0
    
    #Split info in that line
    columns = line.split(' ')
    
    #Remove any empty info in that array
    while '' in columns:
        columns.remove('')
        
    #Assign info
    while col < 14:
        new_table1.iat[row, col] = columns[col]
        col += 1
    row += 1
```

```python
#-------------- FOURTH STEP ----------------------


# Replace all '----' with 'NaN' and we make replace two times bec it hold two values 
new_table1 = new_table1.replace('----', 'NaN').replace('-----', 'NaN')

#check Halo to return True or False
new_table1['is_halo'] = new_table1['cme_angle'].map(lambda x: x == 'Halo')

#Replace "Halo" word with 'NA' word.
new_table1 = new_table1.replace('Halo', 'NA')


#remove any non-numeric part of the width column.
def isInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False
    
#check cme_width with '>'
new_table1['width_lower_bound'] = new_table1['cme_width'].map(lambda x: str(x)[0] == ">")
#then delete the '>'
new_table1['cme_width'] = new_table1['cme_width'].map(lambda x: x if isInt(x) else "")

for row in range(0,len(lines)):
    AllArr = new_table1.values[row][0].split('/')
    tempYear = AllArr[0]
    day_month = new_table1.values[row][2]
    new_table1.iat[row, 2] = tempYear + '/' + day_month
```

```python
# Combine the date column with the time columns to convert to datetime format
new_table1['start_datetime'] = pd.to_datetime(new_table1['start_date'] + '/' + new_table1['start_time'])
new_table1['end_datetime'] = ""
```

```python
# Drop the unneeded columns
new_table1 = new_table1.drop(['end_date', 'end_time'], axis=1)
new_table1
```

```python
#--------------------------- ANALYSIS STEPS -----------------------------
#--------------------------- Question 1: Replication --------------------

# Get the rows that have X class flare
flarex = new_table1['flare_classification'].str.contains('X')

topFlares = new_table1.loc[flarex]
topFlares.is_copy = False

# delete X from the string and we cant use drop bec it will drop the whole string
topFlares['flare_classification'] = topFlares['flare_classification'].str.lstrip('X')

# Change type to float
topFlares['flare_classification'] = topFlares.flare_classification.astype(float)

# Sort rows by flare class
topFlares = topFlares.sort_values('flare_classification', ascending = False)

# Extract the top 50 values of topFlares
topFlares = topFlares.head(50)
# Put back the X in the flare_classification column values
topFlares['flare_classification'] = topFlares.flare_classification.astype(str)
topFlares['flare_classification'] = "X" + topFlares['flare_classification']
topFlares
```
